```python
def dunn_index(data, labels):
    # Küme merkezlerini bul
    centroids = [np.mean(data[labels == label], axis=0) for label in np.unique(labels)]
    
    # Tüm küme merkezlerinin birbirlerine olan uzaklıklarını hesapla
    cluster_distances = pdist(centroids)
    
    # Her bir kümenin çapını bul
    cluster_diameters = [np.max(pdist(data[labels == label])) for label in np.unique(labels)]
    
    # Dunn endeksini hesapla
    dunn_index = np.min(cluster_distances) / np.max(cluster_diameters)
    
    return dunn_index




K = range(2, 13)  # Küme sayıları

# Her bir küme sayısı için Dunn Endeksini hesapla
for k in K:
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(model1)
    labels = kmeans.labels_
    dunn_score = dunn_index(model1, labels)
    print(f"Küme Sayısı: {k}, Dunn Endeksi: {dunn_score}")
```
