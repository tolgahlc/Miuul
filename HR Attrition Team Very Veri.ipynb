```python
def dunn_index(data, labels):
    # KÃ¼me merkezlerini bul
    centroids = [np.mean(data[labels == label], axis=0) for label in np.unique(labels)]
    
    # TÃ¼m kÃ¼me merkezlerinin birbirlerine olan uzaklÄ±klarÄ±nÄ± hesapla
    cluster_distances = pdist(centroids)
    
    # Her bir kÃ¼menin Ã§apÄ±nÄ± bul
    cluster_diameters = [np.max(pdist(data[labels == label])) for label in np.unique(labels)]
    
    # Dunn endeksini hesapla
    dunn_index = np.min(cluster_distances) / np.max(cluster_diameters)
    
    return dunn_index




K = range(2, 13)  # KÃ¼me sayÄ±larÄ±

# Her bir kÃ¼me sayÄ±sÄ± iÃ§in Dunn Endeksini hesapla
for k in K:
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(model1)
    labels = kmeans.labels_
    dunn_score = dunn_index(model1, labels)
    print(f"KÃ¼me SayÄ±sÄ±: {k}, Dunn Endeksi: {dunn_score}")
```
